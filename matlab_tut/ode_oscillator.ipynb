
# <span style="color:rgb(213,80,0)">**Driven Damped Oscillator**</span>

Solving equation of motion for a mass\-spring\-damper system subjected to an external, time\-dependent force $F\left(t\right)$:


$$F\left(t\right)=\;m\frac{d^2 x}{{\textrm{dt}}^2 }+c\frac{\textrm{dx}}{\textrm{dt}}+\textrm{kx}=5\;\sin \left(\omega_f t\right)$$

where:

-  *x* is the displacement (m)
-  m is the mass (2 kg)
-  c is the damping coefficient (0.5 Nm)
-  k is the spring constant (10 N/m)
-  F(t) is the sinusoidal driving force, with ωf�=2 rad/s.

**Initial Conditions:**

-  Initial displacement, x(0)=0 m
-  Initial velocity, dx/dt�(0)=0 m/s

**Time Span:** t∈\[0,50\] seconds.


```matlab
clc; clearvars; close all;

m = 2;
c = 0.5;
k = 10;
w = 2;
tspan = [0 50];
x0 = [0 0];

function dxdt = f(t,x,m,c,k,w)
    dxdt = [x(2); 1/m*(5*sin(w*t)-c*x(2)-k*x(1))];
end

[t,x] = ode45(@(t,x) f(t,x,m,c,k,w), tspan, x0);

plot(t,x, 'LineWidth', 1.5);
legend('Position (x)', 'Velocity (dx/dt)');
xlabel('Time')
ylabel('Value')
```
