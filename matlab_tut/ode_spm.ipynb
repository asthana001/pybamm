
# <span style="color:rgb(213,80,0)">**Single Particle Model (SPM) \- Diffusion**</span>

Modeling the change in **average lithium concentration** ( ${\bar{c} }_s$ ) within the spherical active material particles of a lithium\-ion battery electrode during constant current discharge.


The governing equation for the average concentration ( ${\bar{c} }_s$ ​) is:


$$\frac{d{\bar{c} }_s }{dt}=-3\frac{j}{\mathrm{R}}$$

where:

-  ${\bar{c} }_s$ is the **average concentration** of lithium in the particle (mol/m\$^3\$).
-  t is time (s).
-  R is the particle **radius** (1×10−6 m).
-  j is the **molar flux** of lithium out of the particle surface.

The  **flux j** is related to the **applied discharge current (I)** and the electrode properties:


$$j=\frac{I}{a\;F\;L}$$


where:

-  I is the constant discharge current
-  a is the specific area of the electrode
-  F is Faraday's constant
-  L is the electrode thickness

**Initial Conditions (for the positive electrode):**

-  ${\bar{c} }_s$ ​(0)=28000 mol/m^3 (Partially lithiated state).

**Parameters:**

-  R=1×10^−6 m
-  I=10 A
-  a=4×10^5 m^\-1
-  F=96485 C/mol
-  L=75×10^−6 m

**Time Span:** t∈\[0,3000\] seconds


```matlab
clc; clearvars; close all;

I = 10;
a = 4 * 10^5;
F = 96485;
L = 75 * 10^(-6);
R = 1 * 10^(-6);
tspan = [0 3000];
c0 = [28000];

function dcdt = f(c,t,I,a,F,L,R)
    dcdt = [-3*(I/(a*F*L))/R];
end

[c,t] = ode45(@(c,t) f(c,t,I,a,F,L,R), tspan, c0);

plot(c,t,"LineWidth",1.5);
xlabel("Time (s)");
ylabel("Avg lithium concentration (mol/m^3)");
```
